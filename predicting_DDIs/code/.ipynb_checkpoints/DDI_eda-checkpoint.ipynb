```python
known_drugs_names = pd.read_csv('../data/cleaned data/known_drugs_names.csv')
known_drugs = pd.read_csv('../data/cleaned data/known_drugs.csv')
drug_lookup = pd.read_csv('../data/cleaned data/all_drugs_lookup.csv')
chem_sim = pd.read_csv('../data/cleaned data/chem_sim_eda.csv')
interactions = pd.read_csv('../data/cleaned data/interactions.csv')
```

## Graph of the interaction network
### A visualisation of the undirected interaction network:
Each node represents a drug and each edge (link) represents the interaction between 2 nodes. It isn't really possible to have a granular perspective of the network, but we can safely say that there is a large amount of overlapping which is what we will be indications of similarity. 

```python
#graph object of the known_drugs that outline all of the DDIs
graph = nx.from_pandas_edgelist(known_drugs, "druga_id", "drugb_id", create_using=nx.Graph())

plt.figure(figsize=(10,10))

pos = nx.random_layout(graph, seed=42)

nx.draw(graph, with_labels=False,  pos = pos, node_size = 10, alpha = 0.2, width = 0.1)

plt.show()
```

We have 1404 nodes that represent 89008 interactions. The aim will be to explore the overlap between the interactions.

```python
print(f'Nodes: {graph.number_of_nodes()}'), print(f'Edges: {graph.number_of_edges()}')
```

The depth of a node is the number of edges from the node to the tree's root node.

```python
depths_graph = nx.shortest_path_length(graph, source=0)
depth = max(depths_graph.values())
diameter = nx.algorithms.distance_measures.diameter(graph)
max_clique_size = nx.algorithms.clique.graph_clique_number(graph)
```

```python
print(depth, diameter, max_clique_size)
```

### Drugs interactions


```python
known_count = known_drugs_names.copy()
known_count['count'] = 1
```

```python
a_count = known_count.groupby('drug_a')['count'].sum().sort_values(ascending = False).to_frame().reset_index()
b_count = known_count.groupby('drug_b')['count'].sum().sort_values(ascending = False).to_frame().reset_index()
```

```python
b_count = b_count.rename(columns = {'drug_b': 'drug_a'})
```

```python
int_count = pd.concat([a_count,b_count], axis = 0, ignore_index = True)
```

```python
#sorting drugs from highest amount of interactions to lowest
int_count = int_count.sort_values(by = 'count', ascending = False)
```

```python
#average amount of interactions per unique drug
int_count['count'].mean()
```

```python
top_10 = int_count.rename(columns = {'drug_a':'drug_name'}).head(10)
top_10
```

**Drugs with the most amount of DDIs**<br>Ozanimod has the highest number of interactions with other drugs, amounting to 552 interactions which is well above the average of 67 interactions. The drugs with the most interactions have a higher risk profile as they are more likely to clash with the intake of other drugs.

```python
plt.figure(figsize=(10,6))
plt.title('Top 10 drugs with the most interactions')
plt.xlabel('Number of Interactions')
plt.ylabel('Drug')
plt.barh(top_10['drug_name'],top_10['count'])
```

### Drugs by type of interaction
It seems that antineplastic and immunomodulating agent interactions are the most common, followed by alimentary tract and metabolism interactions. This is an indication that drugs in this area may be more prone to being problematic. It is important to note that DDIs are only confirmed due to research, meaning that this count may be biased towards the most researched fields, and vice versa, where the less prevalent types of interactions may hold many undiscovered interactions.

```python
#finding the sum of each type of interaction
known_count.groupby('type').agg(interactions_sum = ('count', 'sum')).sort_values(by = 'interactions_sum', ascending = False)
```

### Most commonly shared DDI type between the drugs
Many drugs are involved in multiple different types of DDIs. A more accurate representation of the prevalence of DDI types is to investigate each drug and find the number of interactions they have had per type of DDI. What is the most common type of DDI that is 'shared' amongst the drugs?

```python
#sum of types of reactions that each drug is involved with
#preparation for merge
druga_types = known_count.groupby(['drug_a','type'])['count'].sum().to_frame().reset_index().rename(columns = {'drug_a':'drug_name'})
drugb_types = known_count.groupby(['drug_b','type'])['count'].sum().to_frame().reset_index().rename(columns = {'drug_b':'drug_name'})
```

```python
drug_types = pd.concat([druga_types, drugb_types]).groupby(['drug_name', 'type'])['count'].sum().reset_index()
```

```python
drug_types.head()
```

**Summary Statistics**

```python
drug_types.describe().T
```

**Profiles of our most interactive drugs**<br>
If we were to categorise these drugs by their most common DDI type, we can also observe a trend in their DDI type profile. Ozanimod, Cyclosporine and Rucaparib are predominantly involved in antineoplastic and immunomodulating interactions, but also seem to have some involvement with all other DDI types, with alimientary tract and metabolism being the second highest. Similarly, Erythromycin and Fluconazole which are characterised by their dematological DDI prevalence, are only involved in the same pool of 2 other DDI types. Across all drugs, they are all involved with alimentary tract and metabolism DDIs. 

```python
top_5_list = list(top_10['drug_name'])[:5]

def drug_profile(df, drug_list):
    for i,drug in enumerate(drug_list):
        new_df = df[df['drug_name'] == drug]
        plt.figure(i)
        plt.title(f'{drug} DDI involvement')
        sns.barplot(x = 'count', y ='type', data = new_df, color = '#69b3a2')
```

```python
drug_profile(drug_types, top_5_list)
```

**Removing outliers just for EDA**<br>
The count of some drugs will skew the relationships we're trying to explore in the commonality of certain types of DDIs. An inter quartile range analysis reveals that any value over 30 should be considered an outlier. We will remove this for evaluation.

```python
#There is are quite a few drugs that have a much higher prevalence in a certain DDI type
plt.figure(figsize = (15,5))
plt.title('Boxplot of DDI type interactions (count) per drug')
sns.boxplot(x = drug_types['count'])
```

```python
#calculating inter quartile range
iqr = drug_types['count'].quantile(0.75) - drug_types['count'].quantile(0.25)
lower_bound =  drug_types['count'].quantile(0.25) - 1.5*iqr
upper_bound = drug_types['count'].quantile(0.25) + 1.5*iqr
```

```python
lower_bound
```

```python
upper_bound
```

```python
#removing the outliers from the upper bound. There aren't any negative values so we don't consider lower_bound
drug_types_out = drug_types[drug_types['count'] <=upper_bound]
```

**The DDI type most commonly shared between the drugs:**<br>
What we find within the violin plot is the prevalence in the types of interactions shared amongst our portfolio of drugs. We can observe that the alimentary tract and metabolism DDIs have the highest mean, meaning that they are the most common type of DDIs shared across all drugs in the portfolio. This supports the fact that a major contributing factor towards DDIs are complications found in the metabolism of the drug when prescribed with another drug. <br>

Another interesting thing to note is the distribution of these DDI types. We can see that the first, second and last instances have a much more spread distribution. The higher prevalence in the upper bound of these distributions indicate that drugs that are involved in these DDIs are more likely to interact with drugs in general, or with each other. 

```python
plt.figure(figsize=(10,5))
ax = sns.catplot(x="count", y="type", kind="violin", aspect=1, height = 8, data=drug_types_out, palette="Greens")
plt.xlabel('Count')
plt.ylabel('DDI Type')
plt.title('Distribution of DDI type interactions (count) per drug')
```

### Chemical Structures 
Investigating the structural complexity of these drugs

```python
chem_sim['token_length'] = chem_sim['tokens'].transform(len)
```

**Summary statistics**<br>
Mean amount of characteristics in each drug was 137, with a maximum reaching 2686.

```python
chem_sim['token_length'].describe().T
```

**Histogram of the number of tokens found in each drug**<br>
The tokens made for each drug represent the characteristics from their chemical structures. Drugs are predominantly containing around 137 characteristics. We find that there are some outliers that reach up to  2686, meaning that they are drugs with a very extensive and complex structure. 

```python
#histogram of the count in chemical structure tokens
plt.figure(figsize = (10,5))
plt.title('Histogram of the token length of each drug')
sns.histplot(x= 'token_length', data = chem_sim, bins = 50, color = '#1f7444')
#horizontal line to indicate mean
plt.axvline(chem_sim['token_length'].mean(), color='k', linestyle='dashed', linewidth=1)
```

**Count vectorizing**<br>
Transforms our chem structures into matrix data, with a feature for each type of token. We find that there are 235 different characteristics of the chemical structure from our pool of drugs

```python
cvec = CountVectorizer()
```

```python
chem_sim.head()
```

```python
#Creating a dataframe from the sparse matrix that is formed from our Tfidf Transformer
cvec_df = pd.DataFrame(cvec.fit_transform(chem_sim['tokens']).todense(), 
                          columns=cvec.get_feature_names_out())
```

```python
#235 seperate features, 235 different elements of the chemical structures
cvec_df.shape
```

```python
cvec_df
```

```python
cvec_df.to_csv('../data/cleaned data/chem_sim_m.csv', index_label = False)
```

**Most common characteristics of drug chemical structures**<br>
It is a given that carbons are the most abundant. They are the foundation of the structural integrity of everything as they readily bond with other atoms. The carbons with annotations e.g. 'c1, c2' are also pretty common which means that the tokenisation has done its job well because they are indications of carbon structures i.e. carbon rings.

```python
plt.figure(figsize = (10,5))
popular_ch = cvec_df.sum().sort_values(ascending = False).head(10).plot(kind = 'barh', color = '#1f7444')
plt.ylabel('Token')
plt.xlabel('Count')
plt.title('The 10 most common elements of our drug chemical structures')
popular_ch
```
